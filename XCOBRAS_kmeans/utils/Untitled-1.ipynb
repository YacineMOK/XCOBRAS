```python
import pandas as pd

names = [
    "glass_all.csv", 
    "wine_all.csv",
    "wisc_all.csv"
]

name = names[1]

PATH_gt = "../results/ground_truth/"+name

PATH_all = "../results/commun_fraction/"+name

feature_names = ['Top-n', 'Budget', 'ARI']

gt = pd.read_csv(PATH_gt, sep=";", header=None, names=feature_names)
gt['Top-n'] = 0

commun_all = pd.read_csv(PATH_all, sep=";", header=None, names=feature_names)

merge_commun_pd = pd.merge(gt, commun_all, 'outer')
merge_commun_pd['Top-n'][merge_commun_pd['Top-n'] == 0] = "Ground Truth"

verbose = True
if verbose:
    print(f"Ground truth length: {len(gt)} | All length: {len(commun_all)}")
    print(f"sum: {len(gt) + len(commun_all)}")
    print(f"len merged: {len(merge_commun_pd)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(18, 4))

results = merge_commun_pd.pivot(index="Top-n", columns="Budget", values="ARI")
ax.hlines([3], xmax=17, xmin=0, colors="Red")

sns.heatmap(results, annot=True, cmap="Blues")
plt.title("COMMUN FRACTION:         "+name[:-8]+str("       - threashold = 50%"))
plt.show()

fig.savefig("../results/commun_fraction/"+name[:-8]+".jpg")
```

# Lime vs Shap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

names = [
    "glass", 
    "wine",
    "wisc"
]

for name in names:

    gt_feature_names = ['budget','strat','number clusters','ARI']
    commun = ['budget','strat', 'top_n', 'number clusters','ARI']

    gt  = pd.read_csv("../results/ground_truth/" +  name  + "_budget180_GT.csv",  sep=",",  index_col=False )
    rnd = pd.read_csv("../results/random/" +  name  + "_budget180_GT.csv",  sep=",",  index_col=False )
    rnd['top_n'] = -1
    gt['top_n'] = 0

    for i in range(1, 20):
        tmp = pd.read_csv("../results/random/" +  name  + "_budget180_Trial_"+ str(i) + ".csv",  sep=",",  index_col=False)
        rnd['number clusters'] += tmp['number clusters']
        rnd['ARI'] += tmp['ARI']

    rnd['number clusters']/=20
    rnd['ARI']/=20

    lime_commun = pd.read_csv("../results/lime/commun_fraction/" +  name  + "_budget180_n"+ str(3)  +"_th50.csv",  sep=",",  index_col=False )
    lime_commun = lime_commun[commun]
    for n in [5, 7, 9, 10]:
        tmp  = pd.read_csv("../results/lime/commun_fraction/" +  name  + "_budget180_n"+ str(n)  +"_th50.csv",  sep=",",  index_col=False )
        tmp = tmp[commun]
        lime_commun = lime_commun.append(tmp , ignore_index=True)

    shap_commun = pd.read_csv("../results/shap/commun_fraction/" +  name  + "_budget180_n"+ str(3)  +"_th50.csv",  sep=",",  index_col=False )
    shap_commun = shap_commun[commun]
    for n in [5, 7, 9, 10]:
        tmp  = pd.read_csv("../results/shap/commun_fraction/" +  name  + "_budget180_n"+ str(n)  +"_th50.csv",  sep=",",  index_col=False )
        tmp = tmp[commun]
        shap_commun = shap_commun.append(tmp , ignore_index=True)

    all_commun = gt.append(rnd, ignore_index=True)
    all_commun = all_commun.append(lime_commun, ignore_index=True)
    all_commun['top_n'][all_commun['top_n'] ==  0] = "Ground Truth"
    all_commun['top_n'][all_commun['top_n'] == -1]= "Random"
    all_commun

    all_commun_shap = gt.append(rnd, ignore_index=True)
    all_commun_shap = all_commun_shap.append(shap_commun, ignore_index=True)
    all_commun_shap['top_n'][all_commun_shap['top_n'] ==  0] = "Ground Truth"
    all_commun_shap['top_n'][all_commun_shap['top_n'] == -1]= "Random"
    all_commun_shap


    results = all_commun.pivot(index="top_n", columns="budget", values="ARI")
    results_shap = all_commun_shap.pivot(index="top_n", columns="budget", values="ARI")

    fig, ax = plt.subplots(figsize=(18, 4)) 
    ax.hlines([5], xmax=18, xmin=0, colors="Red")
    sns.heatmap(results, annot=True, cmap="Blues")
    plt.title("COMMUN FRACTION:         "+name+str("       - threashold = 50%"))
    plt.show()
    fig.savefig("../results/figures/LIME heatmap commun_fraction_"+name+".jpg")

    fig, ax = plt.subplots(figsize=(18, 4)) 
    ax.hlines([5], xmax=18, xmin=0, colors="Red")
    sns.heatmap(results_shap, annot=True, cmap="Purples")
    plt.title("COMMUN FRACTION:         "+name+str("       - threashold = 50%"))
    plt.show()
    fig.savefig("../results/figures/SHAP heatmap commun_fraction_"+name+".jpg")
```

```python
fig, ax = plt.subplots(figsize=(10, 6)) 
for value in set(all_commun['top_n'].values):
    mask = all_commun['top_n'] == value
    if value =="Ground Truth":
        ax.plot(all_commun['budget'][mask], all_commun['ARI'][mask], linewidth=3, label=str(value))
    else:
        ax.plot(all_commun['budget'][mask], all_commun['ARI'][mask], label=str(value))
    ax.legend()
    fig.savefig("../results/figures/LIME line plot commun_fraction_"+name+"_with random.jpg")

fig, ax = plt.subplots(figsize=(10, 6)) 
for value in set(all_commun['top_n'].values):
    if value != "Random": 
        mask = all_commun['top_n'] == value
        if value =="Ground Truth":
            ax.plot(all_commun['budget'][mask], all_commun['ARI'][mask], linewidth=3, label=str(value))
        else:
            ax.plot(all_commun['budget'][mask], all_commun['ARI'][mask], label=str(value))
        ax.legend()

    fig.savefig("../results/figures/LIME line plot commun_fraction_"+name+"_wo random.jpg")
    
```
