```python
import pandas as pd

names = [
    "glass_all.csv", 
    "wine_all.csv",
    "wisc_all.csv"
]

name = names[0]

PATH_gt = "../results/ground_truth/"+name

PATH_all = "../results/cosine_similarity/"+name

feature_names = ['Top_n', 'Budget', 'ARI']

gt = pd.read_csv(PATH_gt, sep=";", header=None, names=feature_names)
gt['Top_n'] = 0

commun_all = pd.read_csv(PATH_all, sep=";", header=None, names=feature_names)

merge_commun_pd = pd.merge(gt, commun_all, 'outer')
merge_commun_pd['Top_n'][merge_commun_pd['Top_n'] == 0] = "Ground Truth"

verbose = True
if verbose:
    print(f"Ground truth length: {len(gt)} | All length: {len(commun_all)}")
    print(f"sum: {len(gt) + len(commun_all)}")
    print(f"len merged: {len(merge_commun_pd)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(18, 4))

results = merge_commun_pd.pivot(index="Top_n", columns="Budget", values="ARI")
# ax.hlines([4], xmax=14, xmin=0, colors="Red")

sns.heatmap(results, annot=True, cmap="Blues")
plt.title(name[:-8]+str(" - \textit{threashold} = 50%"))
plt.show()
plt.savefig("../results/commun_fraction/"+name[:-8]+".jpg")
```
